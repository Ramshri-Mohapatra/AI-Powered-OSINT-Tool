```python
import torch
import json
import spacy
import numpy as np
from datasets import Dataset, DatasetDict
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForTokenClassification, TrainingArguments, Trainer, DataCollatorForTokenClassification
from sklearn.model_selection import train_test_split
from seqeval.metrics import classification_report
from transformers import EarlyStoppingCallback


#model_name = "distilbert-base-cased" with bnspa and newapi data

# Verify environment
print("Torch version:", torch.__version__)
#print("Transformers version:", transformers.__version__)

# Load spaCy for word-level tokenization
nlp = spacy.load("en_core_web_sm")

# Load the labeled data
bnsapa_dataset = load_dataset("bnsapa/cybersecurity-ner")
with open('/content/sample_data/NewAPI_annotated.json', 'r') as f:
    data1 = json.load(f)
with open('/content/sample_data/Reddit_Twitter_RSS_annotaed.json', 'r') as f:
    data2 = json.load(f)
with open('/content/sample_data/ioc_converted_to_labelstudio.json', 'r') as f:
    ioc_samples = json.load(f)
data3 = data1 + data2

# Debug JSON structure
print("Sample JSON entries:", data3[:2])  # Inspect first two entries

from tqdm import tqdm  # Optional: for nicer progress

print("🔁 Loading IOC examples from Label Studio format...")

ioc_examples = []
for task in tqdm(ioc_samples):
    try:
        text = task.get('data', {}).get('text', '')
        annotations = task.get('annotations', [])
        if not text or not annotations or not annotations[0].get('result'):
            continue

        results = annotations[0]['result']
        doc = nlp(text)
        tokens = [token.text for token in doc]
        token_start_end = [(token.idx, token.idx + len(token.text)) for token in doc]
        labels = ['O'] * len(tokens)

        for ent in results:
            try:
                label = ent['value']['labels'][0]
                start = ent['value']['start']
                end = ent['value']['end']
                for i, (tok_start, tok_end) in enumerate(token_start_end):
                    if tok_start >= start and tok_end <= end:
                        if i == 0 or labels[i-1] == 'O' or labels[i-1][2:] != label:
                            labels[i] = 'B-' + label
                        else:
                            labels[i] = 'I-' + label
            except Exception as inner_e:
                print("⚠️ Annotation issue in result:", inner_e)
                continue

        ner_tags = [label_to_id.get(l, 0) for l in labels]
        ioc_examples.append({'tokens': tokens, 'ner_tags': ner_tags})

    except Exception as outer_e:
        print(f"⚠️ Skipping malformed IOC sample due to error: {outer_e}")
        continue

print(f"✅ Successfully loaded {len(ioc_examples)} IOC examples.")

# Extract text and annotations, converting to BIO format
examples = []
for task in data3:
    text = task.get('data', {}).get('text', '')
    if not isinstance(text, str):
        print(f"Warning: Invalid text in task: {task}")
        continue
    annotations = task.get('annotations', [])
    if not annotations or annotations[0].get('was_cancelled', True):
        print(f"Warning: No valid annotations in task: {task['data']['text'][:50]}...")
        continue
    results = annotations[0].get('result', [])
    if not results:
        print(f"Warning: No results in task: {task['data']['text'][:50]}...")
        continue

    # Tokenize text
    doc = nlp(text)
    tokens = [token.text for token in doc]
    token_start_end = [(token.idx, token.idx + len(token.text)) for token in doc]
    labels = ['O'] * len(tokens)

    # Assign BIO labels
    for ent in results:
        try:
            label = ent['value']['labels'][0]
            start, end = ent['value']['start'], ent['value']['end']
            for i, (tok_start, tok_end) in enumerate(token_start_end):
                if tok_start >= start and tok_end <= end:
                    if i == 0 or labels[i-1] == 'O' or labels[i-1][2:] != label:
                        labels[i] = 'B-' + label
                    else:
                        labels[i] = 'I-' + label
        except (KeyError, IndexError) as e:
            print(f"Warning: Invalid annotation in task: {task['data']['text'][:50]}... Error: {e}")
            continue

    # Define label list
    label_list = ['O', 'B-Organization', 'I-Organization', 'B-System', 'I-System',
                  'B-Malware', 'I-Malware', 'B-Indicator', 'I-Indicator',
                  'B-Vulnerability', 'I-Vulnerability']
    label_to_id = {label: i for i, label in enumerate(label_list)}
    ner_tags = [label_to_id.get(label, 0) for label in labels]

    # Create example dictionary
    example = {'tokens': tokens, 'ner_tags': ner_tags}

    # Debug example
    print(f"Created example: tokens={len(tokens)}, ner_tags={len(ner_tags)}")
    examples.append(example)

# Validate examples
for i, ex in enumerate(examples):
    if not isinstance(ex, dict) or 'tokens' not in ex or 'ner_tags' not in ex:
        print(f"Error: Invalid example at index {i}: {ex}")
        examples[i] = None
examples = [ex for ex in examples if ex is not None]
print("Total valid examples:", len(examples))



all_examples = examples + ioc_examples


# Create Dataset
if not all_examples:
    raise ValueError("No valid examples to create dataset")
dataset = Dataset.from_list(all_examples)
print("Type of dataset:", type(dataset))
print("Dataset size:", len(dataset))
print("First entry:", dataset[0])

from datasets import Sequence, ClassLabel

# Get the ClassLabel object from bnsapa
class_label = bnsapa_dataset["train"].features["ner_tags"].feature

# Cast your dataset's ner_tags column to use the same ClassLabel
dataset = dataset.cast_column("ner_tags", Sequence(feature=class_label))


# Debug dataset

for i, example in enumerate(dataset.select(range(3))):
    print(f"Sample {i}: tokens={len(example['tokens'])}, ner_tags={example['ner_tags']}")
    assert len(example['tokens']) == len(example['ner_tags']), f"Mismatch at index {i}"

# Check label distribution
from collections import Counter
all_labels = [label for example in dataset for label in example['ner_tags']]
label_counts = Counter(all_labels)
print("Label distribution:", {label_list[k]: v for k, v in label_counts.items()})

from datasets import concatenate_datasets

# Combine Hugging Face dataset and your 100 examples
merged_dataset = concatenate_datasets([bnsapa_dataset["train"],dataset ])  # order doesn't matter if you shuffle
merged_dataset = merged_dataset.shuffle(seed=42)

# Then do train-test split
#train_test_split = merged_dataset.train_test_split(test_size=0.2, seed=42) # Used for fine_tuned_model_final1
#train_test_split = bnsapa_dataset["train"].train_test_split(test_size=0.2, seed=42)# Used for fine_tuned_model_final3
train_test_split = dataset.train_test_split(test_size=0.2, seed=42) # Used for fine_tuned_model_final2

train_dataset = train_test_split['train']
val_dataset = train_test_split['test']

# Load pre-trained model and tokenizer
#model_name = "bnsapa/cybersecurity-ner"
model_name = "microsoft/deberta-v3-large"
tokenizer1 = AutoTokenizer.from_pretrained(model_name)
model1 = AutoModelForTokenClassification.from_pretrained(model_name, num_labels=len(label_list))
#model = AutoModelForTokenClassification.from_pretrained("distilbert-base-cased", num_labels=len(label_list))
# Set up label mappings
model1.config.id2label = {i: label for i, label in enumerate(label_list)}
model1.config.label2id = label_to_id

# Tokenize and align labels
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer1(
        examples["tokens"],
        truncation=True,
        padding=True,
        is_split_into_words=True,
    )

    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx])
            else:
                label_ids.append(-100)
            previous_word_idx = word_idx
        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs



# Apply tokenization
tokenized_train = train_dataset.map(tokenize_and_align_labels, batched=True, remove_columns=['tokens', 'ner_tags'])
tokenized_val = val_dataset.map(tokenize_and_align_labels, batched=True, remove_columns=['tokens', 'ner_tags'])

# Set up data collator
data_collator = DataCollatorForTokenClassification(tokenizer1, padding=True)

# Compute metrics for NER
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    true_labels = [[label_list[l] for l in label if l != -100] for label in labels]
    true_predictions = [
        [label_list[p] for (p, l) in zip(pred, label) if l != -100]
        for pred, label in zip(predictions, labels)
    ]

    results = classification_report(true_labels, true_predictions, output_dict=True)
    print("Classification report:", results)
    return {
        "precision": results["weighted avg"]["precision"],
        "recall": results["weighted avg"]["recall"],
        "f1": results["weighted avg"]["f1-score"]
    }

# Set up training arguments
from transformers import TrainingArguments




print("Number of labels in model:", model1.config.num_labels)

for idx, ex in enumerate(dataset):
    for tag in ex["ner_tags"]:
        if tag < 0 or tag >= 11:
            print(f"🚨 Invalid tag {tag} in example {idx}: {ex['tokens']}")


```

```python
from transformers import TrainingArguments, Trainer
from transformers import AutoModelForTokenClassification

training_args_fast = TrainingArguments(
    output_dir="./fast_finetune_results",
    num_train_epochs=15,                        # ✅ Quick polish
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=4,             # Simulate batch size 16
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    weight_decay=0.01,
    save_total_limit=2,
    logging_dir="./fast_finetune_logs",
    logging_steps=25,
    fp16=True,
    gradient_checkpointing=True,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    report_to="none"
)

trainer_fast = Trainer(
    model=model1,
    args=training_args_fast,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    tokenizer=tokenizer1,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# 🚀 Train the model
trainer_fast.train()

# 🧪 Evaluate
eval_results = trainer_fast.evaluate()
print("🧪 Final evaluation results:", eval_results)

# 💾 Save
trainer_fast.save_model('./fine_tuned_model_final1')
tokenizer1.save_pretrained('./fine_tuned_model_final1')

# 🔍 Best checkpoint
print("✅ Best model checkpoint:", trainer_fast.state.best_model_checkpoint)
```

```python
#Works best for my need.
import torch
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
tokenizer_11 = AutoTokenizer.from_pretrained("./fine_tuned_model_final2")
model_11 = AutoModelForTokenClassification.from_pretrained("./fine_tuned_model_final2")  # NewsAPI model

model_11.eval()


# Labels used during training
label_list = ['O', 'B-Organization', 'I-Organization', 'B-System', 'I-System',
              'B-Malware', 'I-Malware', 'B-Indicator', 'I-Indicator',
              'B-Vulnerability', 'I-Vulnerability']


ner = pipeline(
    "ner",
    model=model_11,
    tokenizer=tokenizer_11,
    aggregation_strategy="max",
    device=0 if torch.cuda.is_available() else -1
)

test_sentences = [
    "The IP address 185.220.101.1 was associated with malicious C2 traffic originating from the Emotet botnet.",
    "Researchers found CVE-2025-2748 was exploited by HELLCAT ransomware via WeTransfer.",
    "Multiple holes: (CVE 2023.1234), [CVE–2023–2345], and CVE-2023-3456 detected.",
    "Email me at admin@example.com. Patch CVE.2025.7777 ASAP!",
    "Get free product click - 192.92.1.00",
    "Upgrade to version 1.2.3.4 — it fixes CVE.2021.0005 and CVE‑2021‑0006."
]

for sentence in test_sentences:
    print(f"\n🔍 Input: {sentence}")
    results = ner(sentence)
    for ent in results:
        print(f" - {ent['entity_group']:>12} | {ent['word']:>25} | Score: {ent['score']:.4f}")
```
